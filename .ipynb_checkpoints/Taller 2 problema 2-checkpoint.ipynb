```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_csv("./bitcoin_2017_to_2023.csv")
```

```python
df_clean = df[["open","high","low","close"]]
df_clean
```

```python
plt.figure(figsize=(32, 256))  
contador = 1



for i, title_x in enumerate(df_clean.columns):
    for j, title_y in enumerate(df_clean.columns): 
        plt.subplot(len(df_clean.columns)*4, int(len(df_clean.columns)/4), contador)
        x = df_clean[title_x]
        y = df_clean[title_y]
        
        # Calcular las medias
        media_x = x.mean()
        media_y = y.mean()
        
        # Crear el gráfico de dispersión
        plt.scatter(x, y, color =(0,0,1,0.01), label='Datos', s=10 )
        plt.scatter(media_x, media_y, color='red', marker='o', label='Media', s=100)  # Punto de la media en rojo

        plt.title(title_x + " / " + title_y)
        contador+=1


plt.show()
```

```python
sns.pairplot(df_clean)
```

```python
cov = df_clean.cov()
cov
```

```python
vector_medias = df_clean.mean()
vector_medias
```

```python
matriz_inversa = np.linalg.inv(cov)
matriz_inversa
```

```python
np.dot(cov,matriz_inversa)
```

```python
z = df_clean - vector_medias
z
```

```python
z_t = z.T
z_t
```

```python
producto_punto = np.dot(z_t,z)
producto_punto
```

```python
cov
```

```python
np.corrcoef(np.array(cov).flatten(), np.array(producto_punto).flatten())[0, 1]
```

Punto c. Grafica de distancias de Mahalanobis

```python
distancias = []


for index, row in df_clean.iterrows():
    diff = row - vector_medias
    mahalanobis_dist = np.sqrt(np.dot(np.dot(diff, matriz_inversa), diff))
    distancias.append(mahalanobis_dist)
    


    
# Graficar las variables dos a dos y colorear según la distancia de Mahalanobis
plt.figure(figsize=(12, 8))

# Gráfico 1: Open vs Close
plt.subplot(2, 2, 1)
plt.scatter(df['open'], df['close'], c=distancias, cmap='viridis', s = 2)
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Distancia de Mahalanobis entre Open y Close')
plt.colorbar(label='Distancia de Mahalanobis')

# Gráfico 2: Open vs Low
plt.subplot(2, 2, 2)
plt.scatter(df['open'], df['low'], c=distancias, cmap='viridis', s = 2)
plt.xlabel('Open')
plt.ylabel('Low')
plt.title('Distancia de Mahalanobis entre Open y Low')
plt.colorbar(label='Distancia de Mahalanobis')

# Gráfico 3: Open vs High
plt.subplot(2, 2, 3)
plt.scatter(df['open'], df['high'], c=distancias, cmap='viridis', s = 2)
plt.xlabel('Open')
plt.ylabel('High')
plt.title('Distancia de Mahalanobis entre Open y High')
plt.colorbar(label='Distancia de Mahalanobis')

# Gráfico 4: Close vs Low
plt.subplot(2, 2, 4)
plt.scatter(df['close'], df['low'], c=distancias, cmap='viridis', s = 2)
plt.xlabel('Close')
plt.ylabel('Low')
plt.title('Distancia de Mahalanobis entre Close y Low')
plt.colorbar(label='Distancia de Mahalanobis')

plt.tight_layout()
plt.show()
    
```

Punto d. Bola con radio 1

```python
alpha_values = [0 if d > 1 else 1 for d in distancias]
    
# Graficar las variables dos a dos y colorear según la distancia de Mahalanobis
plt.figure(figsize=(12, 8))

# Gráfico 1: Open vs Close
plt.subplot(2, 2, 1)
plt.scatter(df['open'], df['close'], c=distancias, cmap='viridis', s = 2, alpha=alpha_values)
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Distancia de Mahalanobis entre Open y Close')
plt.colorbar(label='Distancia de Mahalanobis')

# Gráfico 2: Open vs Low
plt.subplot(2, 2, 2)
plt.scatter(df['open'], df['low'], c=distancias, cmap='viridis', s = 2, alpha=alpha_values)
plt.xlabel('Open')
plt.ylabel('Low')
plt.title('Distancia de Mahalanobis entre Open y Low')
plt.colorbar(label='Distancia de Mahalanobis')

# Gráfico 3: Open vs High
plt.subplot(2, 2, 3)
plt.scatter(df['open'], df['high'], c=distancias, cmap='viridis', s = 2, alpha=alpha_values)
plt.xlabel('Open')
plt.ylabel('High')
plt.title('Distancia de Mahalanobis entre Open y High')
plt.colorbar(label='Distancia de Mahalanobis')

# Gráfico 4: Close vs Low
plt.subplot(2, 2, 4)
plt.scatter(df['close'], df['low'], c=distancias, cmap='viridis', s = 2, alpha=alpha_values)
plt.xlabel('Close')
plt.ylabel('Low')
plt.title('Distancia de Mahalanobis entre Close y Low')
plt.colorbar(label='Distancia de Mahalanobis')

plt.tight_layout()
plt.show()
```
